```python
import numpy as np

x = float('nan')      # NaN = Not a Number (résultat indéfini)
y = float('inf')      # inf = infini positif

print(np.isnan(x))    # True  — teste si x est NaN
print(np.isinf(y))    # True  — teste si y est infini
print(x == x)         # False — NaN n'est jamais égal à lui-même !
print(np.nan_to_num(x))  # 0.0 — remplace NaN par 0

print(1.0 / 0.0)     # ZeroDivisionError — mais np.float64(1.0)/0.0 donne inf
```
**Explication :**
- `float('nan')` : crée une valeur *Not a Number*, typique après une division 0/0.
- `np.isnan()` / `np.isinf()` : seules façons fiables de tester ces valeurs (le test `==` échoue pour NaN).

- Méthodes d'algèbre linéaire via `scipy.linalg` : résolution de systèmes ($Ax = b$), décomposition LU, SVD

```python
from scipy.linalg import solve, det
import numpy as np

# Résoudre Ax = b
A = np.array([[2., 1.],
              [5., 7.]])
b = np.array([11., 13.])

x = solve(A, b)   # résolution exacte, plus stable que np.linalg.inv(A) @ b
print('Solution x =', x)
print('Vérification Ax =', A @ x)   # doit valoir [11, 13]
print('Déterminant =', det(A))       # ≠ 0 ⇒ système inversible
```
explication:
- `solve(A, b)` : résout $Ax = b$ par factorisation LU — **plus précis et rapide** que calculer $A^{-1}$.
- `A @ x` : produit matriciel (opérateur `@` introduit en Python 3.5).
- `det(A)` : déterminant — si proche de 0, le système est mal conditionné.

### Chapitre 6 — Visualisation scientifique avec Matplotlib

La visualisation est abordée comme une **compétence scientifique à part entière**, indispensable pour la communication des résultats.

| Sujet | Outils couverts |

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 300)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))  # 1 ligne, 2 colonnes de graphiques

# Premier graphique
axes[0].plot(x, np.sin(x), 'b-', linewidth=2, label=r'$\sin(x)$')
axes[0].plot(x, np.cos(x), 'r--', label=r'$\cos(x)$')
axes[0].set_title('Fonctions trigonométriques')
axes[0].set_xlabel('x'); axes[0].set_ylabel('y')
axes[0].legend()      # affiche la légende
axes[0].grid(True)    # grille

# Deuxième graphique
axes[1].fill_between(x, np.sin(x), 0, alpha=0.3, color='green')
axes[1].set_title('Aire sous sin(x)')

plt.tight_layout()   # espacement automatique
plt.show()
```
explication:
- `plt.subplots(1, 2)` : crée une figure avec 2 sous-graphiques côte à côte ; retourne `fig` et un tableau `axes`.
- `axes[0].plot(x, y, 'b-')` : trace y vs x en bleu (`b`) trait continu (`-`).
- `r'$\sin(x)$'` : chaîne brute avec LaTeX pour la légende.
- `fill_between(x, y, 0)` : remplit l'aire entre la courbe y et la ligne y=0.
- `alpha=0.3` : transparence (0 = invisible, 1 = opaque).
- `tight_layout()` : évite le chevauchement des titres et étiquettes.

Définition de classes, méthodes spéciales (`__init__`, `__add__`, `__repr__`), propriétés (`@property`), méthodes de classe, héritage et encapsulation. Illustrés par une classe `RationalNumber` et une hiérarchie de solveurs d'EDO.

```python
from math import gcd

class RationalNumber:
    """Représente un nombre rationnel p/q en forme réduite."""
    def __init__(self, p, q):
        g = gcd(abs(p), abs(q))   # plus grand commun diviseur
        self.p = p // g           # numérateur réduit
        self.q = q // g           # dénominateur réduit

    def __add__(self, other):     # surcharge de +
        return RationalNumber(self.p * other.q + other.p * self.q,
                               self.q * other.q)

    def __mul__(self, other):     # surcharge de *
        return RationalNumber(self.p * other.p, self.q * other.q)

    def __repr__(self):           # représentation lisible
        return f'{self.p}/{self.q}'

a = RationalNumber(1, 3)
b = RationalNumber(1, 6)
print(a + b)   # 1/2
print(a * b)   # 1/18
```
explication:
- `__init__` : constructeur appelé à la création de l'objet.
- `gcd(p, q)` : réduit la fraction automatiquement (ex : 2/4 → 1/2).
- `__add__` : appelée quand on écrit `a + b` — Python traduit cela en `a.__add__(b)`.
- `__repr__` : définit l'affichage de l'objet (ce que `print` montre).
- Ce patron est au cœur de NumPy : les tableaux utilisent `__add__`, `__mul__`, etc.

```python
def arithmetic_geometric_mean(a, b):
    while abs(a - b) > 1e-12:
        a, b = (a + b) / 2, (a * b) ** 0.5
    yield a
```

### Chapitre 11 — Interfaces graphiques interactives

Programmation de GUI au sein de **Matplotlib** : widgets interactifs (curseurs *sliders*, boutons, zones de texte) connectés à des fonctions de rappel (*callbacks*) déclenchées par des événements utilisateur.

**Nous aborderons les principaux sujets suivants dans ce chapitre :**

* **Un exemple introductif sur les widgets**
* **Le widget bouton et les événements de la souris** (ou **les événements liés à la souris**)

Slider interactif
```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

fig = plt.figure(figsize=(6, 4))

# Axe pour le slider (position dans la figure : [gauche, bas, largeur, hauteur])
sld_ax = plt.axes([0.2, 0.90, 0.6, 0.05])  # bande horizontale en haut

# Axe pour le graphique
ax = plt.axes([0.1, 0.15, 0.8, 0.65])      # zone principale du tracé

# Slider : fréquence ω entre 1 et 5 Hz
sld = Slider(sld_ax,
             label=r'$\omega$ [Hz]',   # étiquette LaTeX
             valmin=1., valmax=5.,      # bornes
             valinit=1.5,              # valeur initiale
             valfmt='%1.1f',           # format d'affichage (1 décimale)
             valstep=0.1)              # pas de déplacement discret

x = np.linspace(-2*np.pi, 2*np.pi, 800)
lines, = ax.plot(x, np.sin(2*np.pi * sld.val * x))

# Fonction de rappel (callback) : appelée à chaque déplacement du slider
def update_frequency(omega):
    lines.set_ydata(np.sin(2*np.pi * omega * x))  # met à jour les données du tracé
    # pas besoin de retracer — set_ydata modifie l'objet graphique en place

sld.on_changed(update_frequency)  # connecte le slider à la fonction
plt.show()
```
- `plt.axes([g, b, l, h])` : positionne manuellement un axe dans la figure en coordonnées relatives [0,1].
- `Slider(ax, label, valmin, valmax, valinit)` : crée un curseur glissant dans l'axe `ax`.
- `sld.val` : valeur courante du slider.
- `lines.set_ydata(...)` : **met à jour les données** d'une courbe existante sans la retracer — bien plus
  efficace que `ax.clear()` + `ax.plot()`.
- `sld.on_changed(callback)` : enregistre la fonction à appeler à chaque changement du slider.
  Le slider passe sa valeur courante comme argument à `callback`.



Bouton et événements souris
```python

from matplotlib.widgets import Button, TextBox
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(7, 4))
plt.subplots_adjust(bottom=0.2)

x = np.linspace(0, 2*np.pi, 400)
line, = ax.plot(x, np.sin(x))

# Bouton 'Update'
btn_ax = plt.axes([0.4, 0.05, 0.2, 0.07])    # zone du bouton
btn = Button(btn_ax, 'Actualiser', hovercolor='salmon')  # rouge au survol

def on_click(event):           # event : objet décrivant le clic
    omega = np.random.uniform(1, 5)
    line.set_ydata(np.sin(omega * x))
    ax.set_title(f'ω = {omega:.2f} Hz')
    fig.canvas.draw_idle()     # redessine la figure de façon asynchrone

btn.on_clicked(on_click)  # on_clicked ≠ on_changed (pour le bouton vs le slider)

# Capture d'un clic souris directement sur la figure
def onclick_figure(event):
    if event.inaxes == ax:     # vérifie que le clic est dans l'axe principal
        print(f'Clic à x={event.xdata:.2f}, y={event.ydata:.2f}')
        if event.button == 3:  # bouton droit de la souris
            ax.plot(event.xdata, event.ydata, 'ro', markersize=8)
            fig.canvas.draw_idle()

# Connexion de l'événement 'button_press_event' à la fonction
cid = fig.canvas.mpl_connect('button_press_event', onclick_figure)
plt.show()
```
- `Button(ax, label, hovercolor)` : widget bouton ; `on_clicked` reçoit un objet `event`.
- `fig.canvas.draw_idle()` : redessine la figure de manière **non bloquante** (contrairement à `plt.draw()`).
- `fig.canvas.mpl_connect('button_press_event', f)` : connecte **tout clic souris** sur la figure
  à la fonction `f` — même sans widget dédié.
- `event.inaxes` : axe dans lequel le clic a eu lieu (`None` si hors des axes).
- `event.button` : numéro du bouton souris (1=gauche, 2=molette, 3=droit).
- `event.xdata`, `event.ydata` : coordonnées du clic dans le **repère du graphique**.

Exceptions personnalisées et gestionnaires de contexte


```python
# --- Exception personnalisée ---
class ConvergenceError(Exception):
    """Levée quand un algorithme itératif ne converge pas."""
    pass

def iterative_solver(tolerance=1e-8, maxiter=1000):
    """Simule un algorithme itératif pouvant échouer."""
    for iteration in range(maxiter):
        residual = 1.0 / (iteration + 1)  # simulation de convergence
        if residual < tolerance:
            return iteration, residual    # succès
    else:
        # Le bloc else s'exécute seulement si la boucle N'a PAS été interrompue par break
        raise ConvergenceError(f"Pas de convergence en {maxiter} itérations")

# Gestion de l'exception
try:
    n_iter, res = iterative_solver(tolerance=1e-4)
    print(f'Convergé en {n_iter} itérations, résidu = {res:.2e}')
except ConvergenceError as e:
    print(f'Erreur : {e}')    # message de l'exception
finally:
    print('Nettoyage toujours exécuté')  # finally : exécuté dans tous les cas

# --- Gestionnaire de contexte ---
with open('/tmp/test_demo.txt', 'w') as f:
    f.write('Données scientifiques\n')   # fichier fermé automatiquement à la sortie du with
print('Fichier créé et fermé automatiquement.')

```
exeplication
- `class ConvergenceError(Exception)` : créer ses propres exceptions permet de distinguer les erreurs
  métier (algorithme) des erreurs Python génériques.
- `for...else` : le `else` d'une boucle `for` est exécuté **seulement si la boucle s'est terminée normalement**
  (sans `break`) — idiome Python peu connu mais très utile pour les algorithmes itératifs.
- `try/except/finally` : `try` tente le code, `except` intercepte les erreurs, `finally` s'exécute toujours
  (nettoyage de ressources).
- `with open(...) as f` : **gestionnaire de contexte** — garantit la fermeture du fichier même en cas d'erreur.

```python
import unittest
import numpy as np

# Fonction à tester
def bisection(f, a, b, tol=1e-10):
    """Méthode de bisection : trouve une racine de f sur [a, b]."""
    if f(a) * f(b) > 0:
        raise ValueError('f doit changer de signe sur [a, b]')
    while (b - a) / 2 > tol:
        mid = (a + b) / 2
        if f(mid) == 0 or (b - a) / 2 < tol:
            return mid
        if f(a) * f(mid) < 0:
            b = mid
        else:
            a = mid
    return (a + b) / 2

# Suite de tests
class TestBisection(unittest.TestCase):

    def test_racine_carre(self):
        """√2 est racine de x²-2 sur [1,2]"""
        root = bisection(lambda x: x**2 - 2, 1, 2)
        self.assertAlmostEqual(root, np.sqrt(2), places=8)  # précision à 8 décimales

    def test_sin(self):
        """π est racine de sin sur [3, 4]"""
        root = bisection(np.sin, 3, 4)
        self.assertAlmostEqual(root, np.pi, places=8)

    def test_signe_incorrect(self):
        """Doit lever ValueError si f ne change pas de signe"""
        with self.assertRaises(ValueError):
            bisection(lambda x: x**2 + 1, 0, 2)  # toujours positif

# Exécution directe dans le notebook
suite = unittest.TestLoader().loadTestsFromTestCase(TestBisection)
runner = unittest.TextTestRunner(verbosity=2)
runner.run(suite)

```
explication:
- `class TestBisection(unittest.TestCase)` : toute classe de test hérite de `TestCase`.
  Chaque méthode commençant par `test_` est un test indépendant.
- `self.assertAlmostEqual(a, b, places=8)` : vérifie que `a` et `b` coïncident à 8 décimales.
  **Indispensable** pour les flottants — `assertEqual` échouerait à cause des erreurs d'arrondi.
- `self.assertRaises(ValueError)` : vérifie qu'une exception spécifique est bien levée.
- `TextTestRunner(verbosity=2)` : affiche un résumé détaillé (OK / FAIL / ERROR pour chaque test).

### Chapitre 16 — Calcul symbolique avec SymPy

**SymPy** est présenté comme un pont entre les mathématiques symboliques et le calcul numérique.

Sujets couverts :
- Symboles et expressions exactes
- Fonctions élémentaires et indéfinies
- Algèbre linéaire symbolique (matrices, valeurs propres)
- Substitution et évaluation numérique via `lambdify`

```python
from sympy import symbols, cos, diff, lambdify, sqrt
import numpy as np

x = symbols('x')
f = cos(x)**2 + sqrt(x)
df = diff(f, x)                     # dérivée symbolique exacte
print(df)                           # -2*cos(x)*sin(x) + 1/(2*sqrt(x))

df_num = lambdify(x, df, 'numpy')   # conversion en fonction NumPy
print(df_num(np.pi/4))              # évaluation numérique
```

Un exemple d'étude de l'**ordre de convergence de la méthode de Newton** illustre la puissance de l'approche hybride symbolique-numérique.

### Chapitre 18 — Calcul parallèle avec mpi4py

L'un des apports majeurs de la 2ᵉ édition : le calcul parallèle sur **architectures multicœurs et clusters** via l'interface MPI (*Message Passing Interface*).

**Nous utiliserons cet exemple comme fil conducteur tout au long de ce chapitre pour aborder les sujets suivants :**

* **Les ordinateurs multicœurs et les grappes de calcul (clusters)**
* **L’interface de passage de messages (MPI – Message Passing Interface)**

| Concept MPI | Description |
|-------------|-------------|
| Rang (*rank*) | Identifiant unique de chaque processus |
| Communication point à point | `send` / `recv` (bloquant et non bloquant) |
| Envoi de tableaux NumPy | `Send` / `Recv` (buffers) |
| `scatter` / `gather` | Distribution et collecte de données |
| `reduce` | Réduction distribuée (somme, max…) |
| `bcast` | Diffusion d'un message à tous les processus |

```python
from mpi4py import MPI
import numpy as np

comm = MPI.COMM_WORLD
rank = comm.Get_rank()
size = comm.Get_size()

# Chaque processus calcule une partie du travail
local_result = rank ** 2
total = comm.reduce(local_result, op=MPI.SUM, root=0)

if rank == 0:
    print(f'Somme des carrés : {total}')  # 0²+1²+…+(size-1)²
```
